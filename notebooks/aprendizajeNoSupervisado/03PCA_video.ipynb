```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
#Tenemos 10 vectores
X_data = np.random.randn(100, 10)
X_data
```

```python
def plot_varianza(prm_pca):
    # Graficar varianza explicada
    plt.figure(figsize=(8, 6))

    # Varianza explicada por cada componente
    plt.subplot(1, 2, 1)
    plt.bar(range(1, len(prm_pca.explained_variance_ratio_) + 1), prm_pca.explained_variance_ratio_)
    plt.xlabel('Componente Principal')
    plt.ylabel('Varianza Explicada')
    plt.title('Varianza Explicada por Componente')

    # Varianza acumulada
    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(prm_pca.explained_variance_ratio_) + 1), prm_pca.explained_variance_ratio_.cumsum(), marker='o', color='r')
    plt.xlabel('Componente Principal')
    plt.ylabel('Varianza Acumulada')
    plt.title('Varianza Acumulada')

    plt.tight_layout()
    plt.show()
```

```python

pca_2components = PCA(n_components=2)

#Tenemos dos vectores
X_pca = pca_2components.fit_transform(X_data)
X_pca


pca_varianza_target = PCA(n_components=.95)

#Tenemos dos vectores
X_pca_varianza = pca_varianza_target.fit_transform(X_data)
X_pca_varianza
```

```python
# Crear subgráficos
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 fila, 2 columnas

# Primer gráfico
axes[0].scatter(X_pca[:, 0], X_pca[:, 1])
axes[0].set_title("Reducción de dimensionalidad con PCA")

# Segundo gráfico
axes[1].scatter(X_pca_varianza[:, 0], X_pca_varianza[:, 1])
axes[1].set_title("Reducción de dimensionalidad con PCA con varianza target 95%")

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

```python
plot_varianza(pca_2components)
plot_varianza(pca_varianza_target)
```
