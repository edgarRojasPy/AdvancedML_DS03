```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
datos = pd.read_csv('celsius.csv')
```

```python
datos.head()
```

```python
X = datos['celsius'].values
y = datos['fahrenheit'].values
```

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),# la entrada, los celsius:1 solo dato
    tf.keras.layers.Dense(units=1, input_shape=[1])# la salida. Tambien es un dato: Los grados fahrenheit
]) 
```

```python
modelo.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
```

```python
entrenamiento = modelo.fit(X, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('# Epochs')
plt.ylabel('Magnitud de la pérdida')
plt.plot(entrenamiento.history["loss"])
```

```python
c = 100
prediccion = modelo.predict(pd.array([c]))
print(prediccion)
```
